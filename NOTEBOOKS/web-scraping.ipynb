```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 

name = []
rating = []
area = []



for page in range(1, 5):
    webPage = requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}".format(page), headers = headers).text
    soup = BeautifulSoup(webPage, "html.parser")
    

    for i in  soup.find_all('div', class_ =  "companyCardWrapper__metaInformation"):
        try:
            name.append(i.find_all('h2', class_ ="companyCardWrapper__companyName")[0].text.strip())
        except:
            name.append(np.nan)
        try:
            rating.append(i.find('div', class_='rating_text--md').text.strip())
        except:
            rating.append(np.nan)
        try:
            area.append(i.find('span', class_ = "companyCardWrapper__interLinking").text.strip())
        except:
            area.append(np.nan)
d = {'name': name, 'rating':rating, 'area': area}
    
print(len(rating))
print(len(name))
print(len(area))
df = pd.DataFrame(d)
df.head()

```
