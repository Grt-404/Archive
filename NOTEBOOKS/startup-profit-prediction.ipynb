# Startup Profit Prediction Analysis

In this analysis, we will predict the profit of startups using a Linear Regression model. We will load a dataset containing information about different startups and their profits.

## Import Necessary Libraries

We begin by importing the required libraries.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Load and Explore the Dataset
We load the dataset '50_Startups.csv' and take a look at its structure

```python
# Load the dataset from a CSV file
df = pd.read_csv('/kaggle/input/50-startups/50_Startups.csv')
df.drop(columns = ['State'], inplace = True)
df =df.iloc[[0,11,36]]


```

```python
df
```

```python
df['f0(x)'] = df['Profit'].mean()
df
```

```python
df['ri1'] = df['Profit'] - df['f0(x)']
df
```

```python
from sklearn.tree import DecisionTreeRegressor
dt1 = DecisionTreeRegressor(max_depth = 1) # as data is small
```

```python
dt1.fit(df.iloc[:, 0:3].values, df['ri1'].values)
```

```python
from sklearn.tree import plot_tree
plt = plot_tree
plt(dt1)

# since we sued mse loss function output of leaves will be equal to output of algo
```
