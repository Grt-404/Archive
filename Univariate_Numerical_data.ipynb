```python
# Mean Median Imputaion 
# if the data is normally distributed use mean(wouldn't matter if you use the mean or median coz the data is normally distributed) and if the data is skewed use median to fill the missing values

# Benefits - 
# Simplicity and thus fast



# Disadvantage - 



# It can distort the original distribution of the data, as many data will now be centric around the mean to median so the estimated variance will decrease so the width of the normal distribution will decrease

# Outliers will come as the IQR is now shrinked and the outliers will be more extreme than before

# jis column pe aap ye lagaoge uska baaki columns ke sath variance might chage or in simple words the covariance will change

# When to use - when the data is missing completely at random (MCAR) and the percentage of missing values is low (typically less than 5%).
```

```python
# Arbitrary value imputation
# if data is not MCAR
# We generally sue this in categorical data where we just say when the data is NA we call it missing
# and in numerical data we can fill it with some value which is out of the range of the data like -9999 or something like that so that the model can learn that this is a missing value and it can learn from it

# Advantages -
# Simple and fast to implement.

# dsdvantages -
# Variance, Covariance changes
# pdf also changes

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='constant', fill_value=-9999)
```

```python
#End Of Distribution Imputation
# this is very much similar to arbitrary imputation but here we fill the missing values with the maximum or minimum value of the column so that the model can learn that this is a missing value and it can learn from it
# if the data is normal we use mean + 3*std to fill the missing values and if the data is skewed we use Q1 - 1.5 IQR or Q3 + 1.5IQR to fill the missing values
```
